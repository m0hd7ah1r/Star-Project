### Basic Cleaning  and Formatting

```python
Movie.drop(["Unnamed: 0"],axis=1,inplace =True) #Dropping the unnamned column
```

```python
#There are Two Column Director and Direcors  lets explore both column and Diretors have only 7 object so focus on Directors
Movie[(Movie["Director"].isnull())&(Movie["Directors"].notnull())][["Director","Directors","Title"]]
```

First let start converting all numerical column into float type and lets Convert into unit of crore

```python
Money_Cols =["Budget","First Day","First Week","First Weekend","India Gross","Worldwide Gross"]

for Cols in Money_Cols:
    Movie[Cols]=Movie[Cols].str.replace(",","").astype('float64')/10000000
```

```python
Movie["Overseas Gross"]=Movie["Overseas Gross"].str.replace(",","").str.replace("$","").astype("float64")/1000000
```

```python
Movie.head()
```

```python
Movie["Runtime"]=Movie["Runtime"].str.replace("min","").astype("float64") #Removing min
```

```python
Movie["Release Date"]=Movie["Release Date"].astype("datetime64[ns]") #Converting to date
```

```python
def plot_barh(X,Y,Data,Title):
    sns.set(style="whitegrid")
    sns.light_palette("navy", reverse=True)
    f, ax = plt.subplots(figsize=(10, 10))
    ax = sns.barplot(x=X, y=Y, data=Data)
    ax.set_xlabel(X)
    ax.set_ylabel(Y)
    cols_title= list(Data[Y])
    cols_gross= list(Data[X])
    i=0
    title_style={'fontsize': 20,
     'fontweight' : 30,
     'verticalalignment': 'baseline',
     'horizontalalignment': "center"}

    plt.title(Title,fontdict=title_style,pad=20,loc="center" )
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.grid(axis=1)
    for p in ax.patches:
        _x = p.get_x()
        _y = p.get_y()+0.05 + (p.get_height()/2)
        value = str(cols_title[i])+"  "+str(round(cols_gross[i],2))+"  Million"
        ax.text(_x, _y, value, ha="left",color="white")
        i=i+1
```

### Some Analysis on Genre of the Movie

```python
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)
labels=list(Movie["Genre"].value_counts().index)
fig, ax = plt.subplots(figsize=(15, 15), subplot_kw=dict(aspect="equal"))
wedges, texts, autotexts = ax.pie(Movie["Genre"].value_counts(), autopct=lambda pct: func(pct,Movie["Genre"].value_counts()),
                                  textprops=dict(color="w"),labels=labels)


ax.legend(wedges,list(Movie["Genre"].value_counts().index),
          title="Genre",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=10, weight="bold")


ax.set_title("India Movie:Genre Distribution",fontdict=title_style,pad=20,loc="center")
plt.axis('equal')

plt.show()
```

### Exploring Actor

```python
Movie["Actors"]= Movie[["Actors 1","Actors 2","Actors 3","Actors 4"]].apply(lambda x:x.tolist(),axis=1)
```

```python
Movie["Actors"].head()
```

```python
Actors=[] #creating a List of All Unique Actors
for row in Movie["Actors"]:
    for actor in row:
        Actors.append(actor)
        
Actors=list(set(Actors))
```

```python
Actors
```

```python
Actors.remove(np.nan)
```

```python
#Lets Create an Actor Tables

ActorList=[]
for actor in Actors:
    ActorsDict={}
    mask = Movie["Actors"].apply(lambda x:actor in x)
    ActorDF = Movie[mask]
    ActorsDict["Actor"]=actor
    ActorsDict["TotalMovie"]=len(ActorDF.index)
    ActorsDict["TotalGross"] = ActorDF["Worldwide Gross"].sum()
    ActorsDict["ROI"] = ActorDF["ROI"].mean()
    ActorList.append(ActorsDict)
    del ActorDF

Actors=pd.DataFrame(ActorList)
```

```python
Actors
```

```python
TopActors = Actors[Actors["TotalMovie"]>10].sort_values("TotalGross",ascending=False)[:20]

TopActors
```

```python
sns.set(style="whitegrid")
sns.light_palette("navy", reverse=True)
f, ax = plt.subplots(figsize=(10, 10))
ax = sns.barplot(x='TotalGross', y='Actor', data=TopActors)
ax.set_xlabel('Total Gross in Crores')
ax.set_ylabel('Actor')
cols_title= list(TopActors["Actor"])
cols_gross= list(TopActors["TotalGross"])
i=0
title_style={'fontsize': 20,
 'fontweight' : 30,
 'verticalalignment': 'baseline',
 'horizontalalignment': "center"}

plt.title("TOP 10 INDIAN ACTOR BY OVERALL GROSS",fontdict=title_style,pad=20,loc="center" )
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.get_yaxis().set_visible(False)
ax.grid(axis=1)
for p in ax.patches:
    _x = p.get_x()+10
    _y = p.get_y() + (p.get_height()/2)
    value = str(cols_title[i])+"  "+str(round(cols_gross[i]))+"  Crore"
    ax.text(_x, _y+0.05, value, ha="left",color="white")
    i=i+1
```
